```python
canadian_youtube = pd.read_csv('./CAvideos.csv')
british_youtube = pd.read_csv('./GBvideos.csv')

canadian_and_british_youtube = pd.concat([canadian_youtube, british_youtube])
canadian_and_british_youtube
```

```python
# Another way to do something similar but removing the repeated trending videos using the video title and trrending date:
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CA', rsuffix='_GB')


'''
The lsuffix and rsuffix parameters are necessary here because the data
has the same column names in both British and Canadian datasets. If this wasn't true (because,
say, we'd renamed them beforehand) we wouldn't need them.
'''
```
